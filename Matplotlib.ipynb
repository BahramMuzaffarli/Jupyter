```python
import matplotlib.pyplot as plt
#from matplotlib import pyplot as plt
```

```python
import numpy as np 
x =np.linspace(0,10,100)
y = np.sin(x)
```

```python
plt.plot(x,np.sin(x))
```

```python
plt.scatter(x,y,color='brown')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(x,0-y,color='b')
plt.plot(x,2-y,color='y')
plt.plot(x,1-y,color='g')
plt.plot(x,np.cos(x),color='black')
```

```python
plt.xlabel('X Burada sizin yaziniz ola bilerdi X',color='r')
plt.ylabel('Y Burada sizin yaziniz ola bilerdi Y',color='r')
plt.show()
```
